```python
plt.plot(x,y,'g',y,x,'b')
plt.xlabel('x Axis Title Here')
plt.ylabel('y Axis Title Here')
plt.title('String Title');
```

```python
plt.subplot(1,2,1)                                        #(1,2,1) 1 for row and 2 for col,
plt.plot(x,y,'y*')
plt.subplot(1,2,2)
plt.plot(x,y,'g--')
```

```python

#object oriented method

```

```python
#create empty canvas
fig=plt.figure()  

#add set of axis                                                                                #( b/w 0 to1)
axes=fig.add_axes([0.1,0.1,0.8,0.8])           #0.1-left,0.1-bottom, 0.8-width, 0.8-height    (range 0 to 1 )
```

```python
axes.plot(x,y,'y')
axes.set_xlabel('x label')                #set is used for to begin methods
axes.set_ylabel('y label')
axes.set_title('Title');
fig
```
